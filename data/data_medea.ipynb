```python
#travel requirement

def visa_formatting(string):
    import re
    string = re.sub("\d","", string)
    string = re.sub(r"[\[\]]", "", string)
    return string
```

```python
# visa requirements status

def visa_status(string):
    import re
    string = string.lower()
    new_string = string.replace("visa not required", "no visa.")
    new_string = new_string.replace("freedom of movement", "no visa.")
    final_string = new_string.split(".",1)
    return final_string[0]
```

```python
## travel requirements for Spanish citizen

url = "https://en.wikipedia.org/wiki/Visa_requirements_for_Spanish_citizens"

travel_spain = pd.read_html(url)[0]
```

```python
travel_spain = travel_spain[["Country", "Visa requirement"]]
```

```python
travel_spain["Visa requirement"] = travel_spain["Visa requirement"].apply(visa_formatting)

travel_spain["Visa requirement"] = travel_spain["Visa requirement"].apply(visa_status)

travel_spain

travel_spain.to_csv("visa_spain.csv", index= False)
```
